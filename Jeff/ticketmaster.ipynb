```python
import requests
import json
from datetime import datetime, timedelta

# Replace with your actual Ticketmaster API key
api_key = 'WckojuMX77PpOVp9exwqrjxtwh6ec1G0'
state_code = 'HI'
country_code = 'US'

# Get today's date and calculate the end date (7 days from today)
today = datetime.utcnow()
end_date = today + timedelta(days=7)

# Format dates in ISO 8601 format
start_date_time = today.strftime('%Y-%m-%dT%H:%M:%SZ')
end_date_time = end_date.strftime('%Y-%m-%dT%H:%M:%SZ')

url = 'https://app.ticketmaster.com/discovery/v2/events.json'

all_events = []
page = 0
size = 50  # Set the desired number of items per page (max is 200 for Ticketmaster API)

while True:
    params = {
        'apikey': api_key,
        'stateCode': state_code,
        'startDateTime': start_date_time,
        'endDateTime': end_date_time,
        'countryCode': country_code,
        'page': page,
        'size': size
    }

    response = requests.get(url, params=params)
    print(response.url)

    if response.status_code == 200:
        events = response.json()
        if '_embedded' in events and 'events' in events['_embedded']:
            event_list = events['_embedded']['events']
            all_events.extend(event_list)
            print(f"Number of events found on page {page}: {len(event_list)}")
        else:
            break  # No more events to fetch
    else:
        print(f'Error: {response.status_code}')
        break

    page += 1

# Print the total number of events retrieved
print(f"Total number of events found: {len(all_events)}")

# Save the JSON response to a file
with open('events.json', 'w') as json_file:
    json.dump({'events': all_events}, json_file, indent=4)
print("JSON response saved to 'events.json'")

# Print event details
for event in all_events:
    print(event.get('name', 'No Name'))
    start_dates = event.get('dates', {}).get('start', {})
    print(start_dates.get('localDate', 'No Date'))
    print(start_dates.get('localTime', 'No Time'))
    classifications = event.get('classifications', [{}])
    genre = classifications[0].get('genre', {})
    print(genre.get('name', 'No Genre'))
```
